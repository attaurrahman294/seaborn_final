# Correlation heatmap


```python
# to check the correlation of each column with different colunn
data.corr()
```

```python
# to  check the correlation of the dataframe on scatter plot
sns.heatmap(data.corr())
plt.show()
```

```python
# to  check the correlation of the dataframe on scatter plot
plt.figure(figsize=(20,8))
sns.heatmap(data.corr() , annot=True, linewidths=4)
plt.show()
```

```python
# to  check the correlation of the dataframe on scatter plot
plt.figure(figsize=(20,8))
sns.heatmap(data.corr() , annot=True, linewidths=4 , vmin=-1 , vmax=1 )
plt.show()
```
